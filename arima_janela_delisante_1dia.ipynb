```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('C:/Users/milen/OneDrive/Documentos/TCC/Bases/clima.csv')

# Convert the Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Forward fill to handle missing values if there's any
df.fillna(method='ffill', inplace=True)

# Function to perform walk-forward validation with EMA
def walk_forward_validation_ema(data, window_size=30):
    predictions = []
    actuals = []
    prediction_dates = []

    total_days = len(data)
    start_date = datetime(2022, 1, 1)
    end_date = datetime(2023, 8, 1)

    while start_date <= end_date:
        train_data = data[(data['Date'] >= start_date - timedelta(days=window_size)) & (data['Date'] < start_date)]['Preco_Real'].values
        val_data = data[data['Date'] == start_date]['Preco_Real'].values

        if len(train_data) > 0 and len(val_data) > 0:
            ema = train_data[-1]  # Initialize EMA with the last value of training data
            alpha = 2 / (window_size + 1)

            for val in train_data[::-1]:
                ema = alpha * val + (1 - alpha) * ema

            predictions.append(ema)
            actuals.append(val_data[0])
            prediction_dates.append(start_date)

        start_date += timedelta(days=1)

    return prediction_dates, actuals, predictions

# Define the window size for the EMA (1 day)
window_size = 30

# Perform walk-forward validation with EMA
prediction_dates, actuals, predictions = walk_forward_validation_ema(df, window_size)

# Calculate error metrics
mse = mean_squared_error(actuals, predictions)
rmse = np.sqrt(mse)
mae = mean_absolute_error(actuals, predictions)
r2 = r2_score(actuals, predictions)

# Output the error metrics
print(f'MSE: {mse}')
print(f'RMSE: {rmse}')
print(f'MAE: {mae}')
print(f'R-squared: {r2}')

# Combine the actual and predicted values for comparison
results = pd.DataFrame({
    'Date': prediction_dates,
    'Actual': actuals,
    'Predicted': predictions
})

# Display the first few rows of the results
print(results.head(10))

# Plot the actual vs. predicted values
plt.figure(figsize=(12, 6))
plt.plot(results['Date'], results['Actual'], label='Actual', linestyle='-', marker='o')
plt.plot(results['Date'], results['Predicted'], label='Predicted', linestyle='-', marker='o')
plt.xlabel('Date')
plt.ylabel('Preco_Real')
plt.title('Actual vs. Predicted Prices')
plt.legend()
plt.grid(True)
plt.show()
```
